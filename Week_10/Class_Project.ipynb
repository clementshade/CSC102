```python
'''
COMMERCIAL BANKING:

Lines of credit(mutual)
Investment management and accounts(mutual)
Insurance(mutual)
Advisory services(mutual)

GLOBAL BANKING:
Multi-currency management services and products(exclusive)
Foreign currency accounts(exclusive)
Foreign currency credit cards(exclusive)
Transborder advisory services(mutual)
Liquidity management(exclusive)

RETAIL BANKING:
Lines of credit(mutual)
Investment management and accounts(mutual)
Insurance(mutual)
Retirement and education accounts(exclusive)
loans and mortgagges(exclusive)
Checking and saving(exclusive)

'''
```

```python
class bank():
    
    def __init__(self,bank_name):
        self.bank_name = bank_name
        self.division_name = None
        self.division = []
        self.mutual_services = []
        self.exclusive_services = []
        
    def add_division(self,x):
        self.division.append(x)
        
    def get_division(self):
        for i in self.division:
            print(i.division_name)
        
    
    def get_mutual_services(self):
        services = ""
        for i in self.mutual_services:
            services += ", "
            services += i
        #does not print mutual services
        if self.mutual_services == []:
            return "None"
        else:
            return f"The mutual services for {self.division_name} are"+services
  
        
    def get_exclusive_services(self):
        services = ""
        #list out the items in the order
        for i in self.exclusive_services:
            services += ", "
            services += i
        #does not print mutual services
        return f"The exclusive services for {self.division_name} are"+services
    


class commercial_banking(bank):  
    
    def __init__(self,bank_name,employee_name):
        self.employee_name = employee_name
        super().__init__(bank_name)
        #super() refers to the "Super class" or "Parent class"
        # for some reason self is not necessary when using super()
        self.division_name = "commercial_banking"
        self.mutual_services = ["Lines of credit", "Investment management and accounts", "Insurance","Advisory services"]


class global_banking(bank):
    def __init__(self,bank_name,employee_name):
        self.employee_name = employee_name
        super().__init__(bank_name)
        self.division_name = "global_banking"
        self.mutual_services = ["Transborder advisory services"]
        self.exclusive_services = ["Multi-currency management services and products","Foreign currency accounts",
                                   "Foreign currency credit cards","Liquidity management"]
    
class retail_banking(bank):
    def __init__(self,bank_name,employee_name):
        self.employee_name = employee_name
        super().__init__(bank_name)
        self.division_name = "retail_banking"
        self.mutual_services = ["Lines of credit", "Investment management and accounts", "Insurance"]
        self.exclusive_services = ["Retirement and education accounts","loans and mortgagges","Checking and saving"]
    

    

# creating bank
zenith = bank("zenith")

#creating division based on employee
Noel = commercial_banking("zenith","Noel")
Agatha = global_banking("zenith","Agatha")
Mary = retail_banking("zenith","Mary")
#adding divisions to bank class 
zenith.add_division(Noel)
zenith.add_division(Agatha)
zenith.add_division(Mary)

def get_services(name):
    is_in_company = False
    for i in zenith.division:
        if name in i.employee_name:
            is_in_company = True
            print("Employee name:",i.employee_name)
            print("Division name:", i.division_name)
            print(i.get_exclusive_services(),end= "\n")
            print(i.get_mutual_services())
        else:
            pass
    if is_in_company == False:
        print("Access Denied")
        
            
    
    pass
#input

ID = str(input("What is your employee name: "))
get_services(ID)
```
