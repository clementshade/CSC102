### Class and Object code

```python

import pandas as pd
df = pd.read_csv('sis.csv')


# Object list for each row in dataframe[Each student's infromation]
stud = []

# Gets specific data for each student
def get_value(x,y):
    # x is the current row in data frame
    # y is the sppecific column
    
    return df.loc[x][y]



class student(object):
    def __init__(self,Mat_no,Name,Age,Grade):
        self.Mat_no = Mat_no
        self.Name = Name
        self.Age = Age
        self.Grade = Grade
        
        
    def info(self):
        """
         Fuction to provide student's information 
        but retuns list of students information instead of printing 
        result of info is a list holding specific informtion
        
        print(f"Mat_no: {self.Mat_no},Name: {self.Name}, Age: {self.Age}, Grade: {self.Grade}")
       
        """
        return [self.Mat_no,self.Name,self.Age,self.Grade]

    
# loops through students info

for i in range(0,len(df.index)):
    x = student(get_value(i,"MatNo"),get_value(i,"Name"),get_value(i,"Age"),get_value(i,"Grade"))
    stud.append(x)
```

### Csv File Code

```python
# Formart for dictionary
stud_dict = {"Mat_no":[],"Name":[],"Age":[],"Grade":[]}
file_name = ""
path = ""
#input
result = int(input("What would do you want to see,\n 1. The Pirates(14-17)\n 2. The Yankees(18-21)\n 3. The Bulls(22-above)\n"))


#functions to set file path
def set_path():
    global path
    path = ("Output/"+file_name+".csv")

#Functions to filter objects by their age
def Pirates(i):
    global file_name
    if i.Age > 13 and i.Age < 18:
        file_name = "the_pirates"
        return i.info()
        
def Yankees(i):
    global file_name
    if i.Age >= 18 and i.Age < 22:
        file_name = "the_yankees"
        return i.info()
def Bulls(i):
    global file_name
    if i.Age >= 22:
        file_name = "the_bulls"
        
        return i.info()
   
        
# loop through Objcect list " stud "
for i in stud:
    
    
    if result == 1:
        choice = Pirates(i)
    elif result == 2:
        choice = Yankees(i)
    elif result == 3:
        choice = Bulls(i)
    else: pass

    
# if the filter function does not return empty add to dataframe dictionary
    if choice != None:
        stud_dict['Mat_no'].append(i.Mat_no)
        stud_dict['Name'].append(i.Name)
        stud_dict['Age'].append(i.Age)
        stud_dict['Grade'].append(i.Grade)

#final path based on choice


#creates dataframe
set_path()
df = pd.DataFrame(stud_dict)

#creates csv file
df.to_csv(path)
print(f"file has been created at {path}")


#display output
df

```
