<p><b>Hints:</b></p>
    <ul><li>The abstract base class is called <b>External_Vendors()</b>.</li><li>
        The abstract method is called <b>menu()</b>.</li><li>
The subclasses (the different vendors) will inherit the abstract base class.</li><li>
        Each subclass will have a normal method called <b>menu()</b>.</li></ul>
    
       

```python
from abc import ABC,abstractmethod
import pandas as pd

class External_Vendors(ABC):
    @abstractmethod
    def menu(self):
        pass
class Cooperative(External_Vendors):
    def __init__(self):
        self.d1 = {"Main Meal":["Jollof Rice and Stew","White Rice and Stew","Fried Rice",
                   "Salad","Platain"],"Price":["200","200","200","100","100"]} 
        self.df = pd.DataFrame(self.d1)
    def menu(self):
        print(self.df)
        pass
class Faith(External_Vendors):
    def __init__(self):
        self.d2 = {"Main Meal":["Fried Rice","White Rice and Stew","Jollof Rice",
                   "Beans","Chicken"],"Price":["400","400","400","200","1000"]} 
        self.de = pd.DataFrame(self.d2)
    def menu(self):
        print(self.de)
        pass
class Student(External_Vendors):
    def __init__(self):
        self.d3 = {"Main Meal":["Chicken Fried Rice","Pomo Sayce","Spaghetti Jollof",
                   "Amala/Ewedu","Semo with Eforiro Soup"],"Price":["800","300","500","500","500"]} 
        self.de = pd.DataFrame(self.d3)
    def menu(self):
        print(self.de)
        pass
def order():
    global x
    x = int(input("""
    Enter number of club:
    1.Cooperative Cafeteria
    2.Faith Hostel Cafeteria
    3.Student Centre Cafeteria

    """))

order()
if x == 1:
    food = Cooperative()
    food.menu()
elif x==2:
    food = Faith()
    food.menu()
elif x==3:
    food = Student()
    food.menu()

else: 
    print("Wrong Input")
```
